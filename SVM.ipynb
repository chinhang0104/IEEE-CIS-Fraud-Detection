```python
X.fillna(-999, inplace=True)
```

```python
from sklearn import svm
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
y.value_counts()
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_sm, y_sm = smote.fit_sample(X, y)
```

```python
y_sm.value_counts()
```

```python
from sklearn import preprocessing
```

```python
x = X_sm.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
X_sm = pd.DataFrame(x_scaled)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_sm, y_sm, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
X_train['y_train'] = y_train
```

```python
X_train
```

```python
X_train = X_train.sample(frac=1).reset_index(drop=True)
```

```python
X_train
```

```python
X_train_sub1 = X_train.iloc[list(range(3000))]
```

```python
y_train_sub1 = X_train_sub1['y_train']
y_train_sub1
```

```python
X_train_sub1.drop(columns=['y_train'], inplace=True)
X_train_sub1
```

```python
#Radial Basis Function
clf=svm.SVC(kernel='rbf',C=1,gamma='auto')
clf.fit(X_train_sub1,y_train_sub1)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
```

```python
y_pred = clf.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
f1_score(y_test, y_pred)
```
